```python
# Necessary Imports

import pandas as pd
from datetime import date, timedelta, datetime
import yfinance as yf
import json
import requests
```

```python
# Generate list with all stocks available in B3

url = "https://brapi.dev/api/available"
params = {"token": "5HNNFP5vqTeSvPHZ4kyaSL"}

def generate_list_stocks(url, params):
    response = requests.get(url, params)
    if response.status_code == 200:
        data = response.json()
        list_stocks = []
        for stock in data["stocks"]:
            list_stocks.append(stock)
        return list_stocks
    else:
        print(f"Error: {response.status_code}")
        return None
    
# def create_date(list_dates):
#     for date in list_dates:
#         date = date.strftime('%Y-%m-%d')
#     return list_dates

# def create_list_dates(data_info, period):
#     list_dates = []
#     stock = 'PETR4'
#     chamada_api = yf.Ticker(f'{stock}.SA').history(period=period)
#     for date in chamada_api[data_info].index:     
#         start_date, end_date = create_date(date)
#         list_dates.append(start_date)
#     return list_dates

def create_list_dates(list_dates):
    formatted_list_dates = []
    for date in list_dates:     
        formatted_date = date.strftime('%Y-%m-%d')
        formatted_list_dates.append(formatted_date)
    return formatted_list_dates

list_stocks = generate_list_stocks(url, params)
```

```python
# df_dividends

df_dividends = pd.DataFrame()
list_dividends = []

for stock in list_stocks:    
    stock_info = yf.Ticker(f'{stock}.SA').history(period="36mo")['Dividends']
    list_dates = create_list_dates(stock_info.index)
    list_dividends = list(stock_info.values)
    
    obj_stock = {
        stock: list_dividends
    }
#    Creation of temporary DataFrame
    try:
        temp_df = pd.DataFrame({
            date: dividend
            for date, dividend in zip(list_dates, obj_stock[stock])
        }, index=[stock])
    except Exception as e:
        print(f'Erro(f{stock}) - {e}')
#     Concatenate temp_df with the df_dividends
    try:
        df_dividends = pd.concat([df_dividends, temp_df], axis=0)
    except Exception as e:
        print(f'Erro: {e}')
        
df_dividends.to_csv("df_dividends.csv", index=True)
```
