```python
#0 FAZER OS IMPORTS NECESSÁRIOS

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import date, timedelta
import yfinance as yf
import json
import requests
```

```python
#1 GERAR LISTA COM O NOME DE TODAS AS AÇÕES

url = "https://brapi.dev/api/available"
params = {"token": "5HNNFP5vqTeSvPHZ4kyaSL"}
response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()
    array_stocks = []
    for stock in data["stocks"]:
        array_stocks.append(stock)
else:
    print(f"Erro {response.status_code}")
```

```python
#2 Criar dataset de Dividends
#2.1 Jogar os dados dos dividendos no df_dividends

df_dividends = pd.DataFrame()

for ativo in array_stocks:
    chamada_api = yf.Ticker(f'{ativo}.SA').history(period="36mo")
    dividends = chamada_api['Dividends'].values
    try:
        df_dividends[ativo] = dividends
    except:
        continue
        
#2.2 Colocar a data como index

date = []

for stock in array_stocks:
    stock_history = yf.Ticker(f'{stock}.SA').history(period="36mo")
    date.append(list(stock_history["Dividends"].index)) 

df_dividends["Date"] = date[array_stocks.index('BHIA3')] # A coluna 'Date' vai receber os valores de data do stock com o maior registro, no caso 'BHIA3'
df_dividends["Date"] = df_dividends["Date"].dt.tz_convert(None) 
df_dividends = df_dividends.set_index("Date", drop=True)

#2.3 Converter para csv

df_dividends.to_csv("df_dividends.csv", index=True)
```
